1. Importación de paquetes y web scraping de OLX Colombia. 

```python
import requests
import pandas as pd
headers = {
    'user_agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'
}

carros_totales = []
for i in range(0, 20):
    url_api = "https://www.olx.com.co/api/relevance/v2/search?category=378&facet_limit=100&location=1000001&location_facet_limit=20&platform=web-desktop&size=40&user=17f0f5ce4eax7b0e8478&page=" + str(i)
    response = requests.get(url_api, headers = headers)
    data = response.json()
    
    autos = data["data"]
    for auto in autos:
        carros_totales.append(auto)

df = pd.DataFrame(carros_totales)
df.info()


    




```
